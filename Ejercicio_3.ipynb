```python
import boto3

# Crear cliente EC2
ec2 = boto3.client('ec2')
```

```python

def listar_instancias():
    try:
        # Obtener información de todas las instancias
        response = ec2.describe_instances()

        # Recorrer reservas e instancias
        for reserva in response['Reservations']:
            for instancia in reserva['Instances']:

                instance_id = instancia.get('InstanceId', 'N/A')
                estado = instancia.get('State', {}).get('Name', 'N/A')
                tipo = instancia.get('InstanceType', 'N/A')

                # Algunas instancias no tienen IP pública
                ip_publica = instancia.get('PublicIpAddress', 'Sin IP pública')

                print("Instance ID:", instance_id)
                print("Estado:", estado)
                print("Tipo:", tipo)
                print("IP pública:", ip_publica)
                print("-" * 40)

    except Exception as e:
        print("Error al obtener instancias:", e)
```

```python
if __name__ == "__main__":
    listar_instancias()
```
