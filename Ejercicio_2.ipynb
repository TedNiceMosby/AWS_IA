```python
import boto3
from datetime import datetime, timezone, timedelta
```

```python
# Crear cliente S3
s3 = boto3.client('s3')

# Nombre del bucket
BUCKET_NAME = "tu-bucket-aqui"

# Días límite para eliminar archivos
DIAS_LIMITE = 30
```
