```python
import boto3

# Crear cliente IAM
iam = boto3.client('iam')
```

```python
def crear_usuario(nombre_usuario):
    try:
        response = iam.create_user(UserName=nombre_usuario)
        print(f"Usuario creado: {nombre_usuario}")
        return response
    except Exception as e:
        print("Error al crear usuario:", e)


def asignar_politica(nombre_usuario, arn_politica):
    try:
        iam.attach_user_policy(
            UserName=nombre_usuario,
            PolicyArn=arn_politica
        )
        print(f"Política asignada a {nombre_usuario}")
    except Exception as e:
        print("Error al asignar política:", e)


def crear_access_keys(nombre_usuario):
    try:
        response = iam.create_access_key(UserName=nombre_usuario)

        access_key = response['AccessKey']['AccessKeyId']
        secret_key = response['AccessKey']['SecretAccessKey']

        print("Access Key creada:")
        print("AccessKeyId:", access_key)
        print("SecretAccessKey:", secret_key)

        return access_key, secret_key

    except Exception as e:
        print("Error al crear access keys:", e)
```

```python
if __name__ == "__main__":
    nombre = input("Nombre del usuario: ")

    # ARN de ejemplo (acceso completo a S3)
    politica = "arn:aws:iam::aws:policy/AmazonS3FullAccess"

    crear_usuario(nombre)
    asignar_politica(nombre, politica)
    crear_access_keys(nombre)
```
