```python
import boto3
import random
import time

# Crear cliente CloudWatch
cloudwatch = boto3.client('cloudwatch')

# Configuración
NAMESPACE = "MiAplicacion"
METRIC_NAME = "UsoCPU"
```

```python

def enviar_metrica(valor):
    try:
        response = cloudwatch.put_metric_data(
            Namespace=NAMESPACE,
            MetricData=[
                {
                    'MetricName': METRIC_NAME,
                    'Value': valor,
                    'Unit': 'Percent'
                }
            ]
        )
        print(f"Métrica enviada: {valor}%")
    except Exception as e:
        print("Error al enviar métrica:", e)
```

```python
if __name__ == "__main__":
    # Simulación de envío de métricas cada 5 segundos
    for _ in range(5):
        valor_cpu = random.randint(0, 100)
        enviar_metrica(valor_cpu)
        time.sleep(5)
```
