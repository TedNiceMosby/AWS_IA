```python
import boto3
```

```python
# Crear cliente S3
s3 = boto3.client('s3')

# 🔹 CONFIGURA TU BUCKET
BUCKET_NAME = "tu-bucket-aqui"
```

```python
# Subir archivo
def subir_archivo(ruta_local, nombre_s3):
    try:
        s3.upload_file(ruta_local, BUCKET_NAME, nombre_s3)
        print(f"Archivo '{nombre_s3}' subido correctamente.")
    except Exception as e:
        print("Error al subir:", e)
```

```python
# Listar los archivos
def listar_archivos():
    try:
        response = s3.list_objects_v2(Bucket=BUCKET_NAME)
        if 'Contents' in response:
            print("Archivos en el bucket:")
            for obj in response['Contents']:
                print("-", obj['Key'])
        else:
            print("El bucket está vacío.")
    except Exception as e:
        print("Error al listar:", e)
```

```python
# Descargar archivo
def descargar_archivo(nombre_s3, ruta_local):
    try:
        s3.download_file(BUCKET_NAME, nombre_s3, ruta_local)
        print(f"Archivo '{nombre_s3}' descargado.")
    except Exception as e:
        print("Error al descargar:", e)
```

```python
# Ejemplo de uso
if __name__ == "__main__":
    print("1. Subir archivo")
    print("2. Listar archivos")
    print("3. Descargar archivo")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        ruta = input("Ruta del archivo local: ")
        nombre = input("Nombre en S3: ")
        subir_archivo(ruta, nombre)

    elif opcion == "2":
        listar_archivos()

    elif opcion == "3":
        nombre = input("Nombre en S3: ")
        ruta = input("Ruta destino: ")
        descargar_archivo(nombre, ruta)
```
